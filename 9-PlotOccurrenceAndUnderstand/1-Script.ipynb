```python
import pandas as pd
```

```python
data = pd.read_csv("../8-UnderstandMetrics/output.csv", index_col=0)
```

```python
data.columns
```

```python
data["percentage_duplicate_lines"] = (data["total_duplicated_lines"]/data["CountTotalLine"]) * 100
data["percentage_duplicate_lines_of_code"] = (data["total_duplicated_lines"]/data["CountTotalLOC"]) * 100
```

```python
data = data[data["percentage_duplicate_lines"] < 100]
data = data[data["percentage_duplicate_lines_of_code"] < 100]
```

```python
for index in data.index:
    print(f"Index {index}")
    serie = data.loc[index]
    if serie["framework"] == "googlearchive":
        data.loc[index, "framework"] = "Android"
    if serie["framework"] == "googlesamples":
        data.loc[index, "framework"] = "Android"
    if serie["framework"] == "spring-guides":
        data.loc[index, "framework"] = "Spring"
```

```python
android = data[data["framework"] == "Android"]
```

```python
spring = data[data["framework"] == "Spring"]
```

```python
android.to_csv("android.csv", index=False)
```

```python
spring.to_csv("spring.csv", index=False)
```
